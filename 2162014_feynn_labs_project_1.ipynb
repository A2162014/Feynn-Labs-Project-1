# Data Analysis and Visualization

```python
# Line plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=df[['V', 'VM']])
plt.title('Line Plot')
plt.xlabel('Index')
plt.ylabel('Values')
plt.show()
```

```python
# Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='T', y='PM 2.5', data=df)
plt.title('Scatter Plot')
plt.xlabel('T')
plt.ylabel('PM 2.5')
plt.show()
```

```python
# Histogram
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='PM 2.5', kde=True)
plt.title('Histogram')
plt.xlabel('PM 2.5')
plt.ylabel('Frequency')
plt.show()
```

```python
# Bar plot
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='SLP')
plt.title('Bar Plot')
plt.xlabel('SLP')
plt.ylabel('Count')
plt.show()
```

```python
# Box plot
plt.figure(figsize=(10, 6))
sns.boxplot(data=df[['T', 'TM', 'Tm']])
plt.title('Box Plot')
plt.ylabel('Values')
plt.show()
```
